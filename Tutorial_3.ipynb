```python
import numpy as np
```

```python
arr_2x2 = np.array([[1, 2],
                    [3, 4]])
print("2x2 array:\n", arr_2x2)
```

```python
rand_3x4 = np.random.rand(3, 4)  # values between 0 and 1
print("Random 3x4 array:\n", rand_3x4)
```

```python
ones_arr = np.ones((2, 3))
zeros_arr = np.zeros((2, 3))

print("Array of ones:\n", ones_arr)
print("Array of zeroes:\n", zeros_arr)
```

```python
rand_6x3 = np.random.rand(6, 3)
print("Original 6x3 array:\n", rand_6x3)

transpose_arr = rand_6x3.T
print("Transposed array (3x6):\n", transpose_arr)
```

```python
A = np.random.rand(2, 3)
B = np.random.rand(3, 4)

dot_product = np.dot(A, B)
print("Matrix A (2x3):\n", A)
print("Matrix B (3x4):\n", B)
print("Dot product (2x4):\n", dot_product)
```
